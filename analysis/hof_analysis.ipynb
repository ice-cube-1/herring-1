```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("output.csv")
prey_cols = ['alpha','beta','gamma','delta','alive']
predator_cols = ['gamma1','gamma2','k','alive']
prey_data = pd.DataFrame(columns=prey_cols)
predator_data = pd.DataFrame(columns=predator_cols)
for i in range(2000):
    prey_data.loc[i] = data.iloc[i*10+5:(i+1)*10][prey_cols].mean()
    predator_data.loc[i] = data.iloc[i*10:i*10+5][predator_cols].mean()
```

```python
plt.plot(prey_data["alive"],label='prey')
plt.plot(predator_data["alive"],label='predator')
plt.legend()
plt.show()
plt.plot(prey_data["alpha"])
plt.plot(prey_data["beta"])
plt.show()
plt.plot(prey_data["delta"])
plt.show()
plt.plot(predator_data["gamma1"],label='g1')
plt.plot(predator_data["gamma2"],label='g2')
plt.legend()
plt.show()
```

```python
print(predator_data[predator_data.index>325].sort_values("alive",ascending=True).head(10))
print(prey_data[prey_data.index>325].sort_values("alive",ascending=False).head(10))
```
