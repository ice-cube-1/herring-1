```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("output.csv")
prey_cols = ['alpha','beta','gamma','delta','alive']
predator_cols = ['gamma1','gamma2','k','alive']
prey_data = pd.DataFrame(columns=prey_cols)
predator_data = pd.DataFrame(columns=predator_cols)
for i in range(2000):
    prey_data.loc[i] = data.iloc[i*10+5:(i+1)*10][prey_cols].mean()
    predator_data.loc[i] = data.iloc[i*10:i*10+5][predator_cols].mean()
```

```python
plt.plot(prey_data["alive"],label='prey')
plt.plot(predator_data["alive"],label='predator')
plt.legend()
plt.show()
plt.scatter(y=prey_data["alpha"],x=prey_data.index,s=2)
plt.scatter(y=prey_data["beta"],x=prey_data.index,s=2)
plt.show()
plt.scatter(y=prey_data["delta"],x=prey_data.index,s=2)
plt.show()
plt.scatter(y=prey_data["gamma"],x=prey_data.index,s=2)
plt.show()
plt.scatter(y=predator_data["gamma1"],x=predator_data.index,label='g1',s=2)
plt.scatter(y=predator_data["gamma2"],x=predator_data.index,label='g2',s=2)
plt.legend()
plt.show()
plt.scatter(y=predator_data["k"],x=predator_data.index,label='g2',s=2)
plt.show()
```

```python
predator_data= predator_data[predator_data.index>200].sort_values("alive",ascending=True)
prey_data = prey_data[prey_data.index>200].sort_values("alive",ascending=False)
print(predator_data.head(10))
print(prey_data.head(10))

```

```python
plt.scatter(prey_data["alpha"],prey_data['alive'],s=2,label='alpha')
plt.scatter(prey_data["beta"],prey_data['alive'],s=2,label='beta')
plt.legend()
plt.show()
plt.scatter(prey_data["alpha"],prey_data['beta'],s=(prey_data['alive']-4500)/4)
plt.show()
plt.scatter(prey_data["delta"],prey_data['alive'],s=2)
plt.show()
plt.scatter(prey_data["gamma"],prey_data['alive'],s=2)
plt.show()
plt.scatter(predator_data["gamma1"],predator_data['gamma2'],c=predator_data['alive'],cmap='coolwarm_r')
plt.show()
plt.scatter(predator_data["gamma1"],predator_data['alive'],label='gamma1',s=2)
plt.scatter(predator_data["gamma2"],predator_data['alive'],label='gamma2',s=2)
plt.legend()
plt.show()
plt.scatter(predator_data["k"],predator_data['alive'],label='g2',s=2)
plt.show()
```
